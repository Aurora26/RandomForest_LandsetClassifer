```python
import scipy.io as sio
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
def load_data(path): 
    csv_path = os.path.join('/Users/aurora/Desktop/CSV/2016/',path)
    return pd.read_csv(csv_path)
```

```python
data_2016=[]
for a,b,c in os.walk('/Users/aurora/Desktop/CSV/2016/'):
    for i in c:
        if (i!='.DS_Store'):
            data = load_data(i)
            data_2016.append(data)
data_2016=pd.concat(data_2016)
data_2016.head(10)
print(len(data_2016))
```

```python
#去除重复数据
data_2016 = data_2016.drop_duplicates()
data_2016 = data_2016.loc[data_2016["band1"]!=65536]
print(len(data_2016))
data_2016.head(10)
```

```python
# 切分test数据集和train数据集
np.random.seed(42)#随机数种子，该值不变每次随机打乱结果都一样
shuffled_indces = np.random.permutation(data_2016.shape[0])#data就是list,随机打乱行，若是data.shape[1]就是打乱列
data_2016 = data_2016.iloc[shuffled_indces]#切分过的训练集list
```

```python
print(len(data_2016))
data_2016.head(10)
```

```python
data = data_2016[:]
zero_loc = data_2016.loc[data_2016["label"]==0]
one_loc = data_2016.loc[data_2016["label"]==1]
two_loc = data_2016.loc[data_2016["label"]==2]
three_loc = data_2016.loc[data_2016["label"]==3]
print(len(zero_loc),len(one_loc),len(two_loc),len(three_loc))
```

```python
real_loc = data.loc[data["band1"]!=32767]
#real_loc = real_loc.drop(['band1','band2','band3','band4','band5','band6','band7'],1)
print(len(real_loc))
real_loc.head(20)
```

```python
zero_loc = data.loc[data["label"]==0]
one_loc = data.loc[data["label"]==1]
two_loc = data.loc[data["label"]==2]
three_loc = data.loc[data["label"]==3]
print(len(zero_loc),len(one_loc),len(two_loc),len(three_loc))
```

```python
zero_loc = zero_loc[0:50000]
zero_loc.head(10)
```

```python
ture_data = [zero_loc,one_loc,two_loc,three_loc]
ture_data = pd.concat(ture_data)#合并
print(len(zero_loc),len(one_loc),len(two_loc),len(three_loc))
ture_data.head(20)
```

```python
# 切分test数据集和train数据集
np.random.seed(42)#随机数种子，该值不变每次随机打乱结果都一样
shuffled_indces = np.random.permutation(ture_data.shape[0])#data就是list,随机打乱行
val_set_size = int(ture_data.shape[0] * 0.2)#这个比例根据需求来
val_indices = shuffled_indces[:val_set_size]
train_indices = shuffled_indces[val_set_size:]
train_data = ture_data.iloc[train_indices]#切分过的训练集list
val_data = ture_data.iloc[val_indices]#切分过的测试集list
```

```python
import scipy.io as sio
import os
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from scipy import misc
from skimage import transform,data
import gdal
from sklearn.ensemble import RandomForestClassifier
from sklearn.datasets import make_classification
import joblib
from sklearn import metrics
from sklearn import preprocessing
import spectral
import sys
```

```python
def load_data(path): 
    csv_path = os.path.join('/Users/aurora/Desktop/CSV/2016/',path)
    return pd.read_csv(csv_path)
```

```python
train_data = load_data('train_data.csv')
test_data = load_data('test_data.csv')
```

```python
train_x = train_data.drop(['label'],1)
train_y = train_data['label']
test_x = test_data.drop(['label'],1)
test_y = test_data['label']
print(len(train_x))
```

# oob

```python
accuracy = []
for i in range(100):
    clf = RandomForestClassifier(n_estimators=i+1,random_state=0,oob_score=True)
    clf.fit(train_x,train_y)
    acc = clf.oob_score_
    sys.stdout.write('\r >>Process %d/%f'%(i+1,acc))
    accuracy.append(np.array([i,acc]))
```

```python
acc=pd.DataFrame(np.array(accuracy))
```

```python
acc.to_csv('/Users/aurora/Desktop/npy/accuracy.csv')
```

```python
fig=plt.figure(figsize=(15,6))
ax = fig.add_subplot(1,1,1)
acc_array = np.array(accuracy)
plt.plot(acc_array[:,0],acc_array[:,1],label="Acc Rate")
ax.set_title('Accuracy')
ax.set_ylabel('Acc rate')
ax.set_xlabel('Tree Num')
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(15,6))
ax = fig.add_subplot(1,1,1)
acc_array = np.array(accuracy)
plt.plot(acc_array[:,0],1-acc_array[:,1],label="oob Rate")
ax.set_title('OOB error')
ax.set_ylabel('oob error rate')
ax.set_xlabel('Tree Num')
plt.legend()
plt.show()
```

# save model

```python
clf = RandomForestClassifier(n_estimators=50,random_state=0,oob_score=True)
clf.fit(train_x,train_y)
```

```python
pred=clf.predict(test_x)
accuracy=metrics.accuracy_score(test_y,pred)*100
print(accuracy)
```

```python
joblib.dump(clf,"randomforest_model2016.m")
```

```python
def show_accuracy(x,y,string):
    acc = list(np.equal(x,y)).count(True)/(len(x))
    print(string,' acc=',' ',acc)
x=clf.predict(train_data.drop(['label'],1))
y=train_y
show_accuracy(x,y,'ceshi')
```

```python
x=clf.predict(test_data.drop(['label'],1))
y=test_y
show_accuracy(x,y,'ceshi')
```

```python
def apply_mask(image, mask, color, mask_num=1,alpha=0.5):
    """Apply the given mask to the image.
    """
    for c in range(3):
        image[:, :, c] = np.where(mask == mask_num,
                                  image[:, :, c] *
                                  (1 - alpha) + alpha * color[c],
                                  image[:, :, c])
    return image
```
